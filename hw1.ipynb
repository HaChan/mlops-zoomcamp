# Question 2 Computing duration
Now let's compute the duration variable. It should contain the duration of a ride in minutes.

What's the standard deviation of the trips duration in January?

- 32.59
- 42.59
- 52.59
- 62.59


```python
df['duration'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60
```

```python
df['duration'].std()
```

## Q3. Dropping outliers

Next, we need to check the distribution of the `duration` variable. There are some outliers. Let's remove them and keep only the records where the duration was between 1 and 60 minutes (inclusive).

What fraction of the records left after you dropped the outliers?

* 90%
* 92%
* 95%
* 98%

```python
df_norm = df[df['duration'].between(1, 60)]
```

```python
df_norm.shape[0] / df.shape[0]
```

## Q4. One-hot encoding

Let's apply one-hot encoding to the pickup and dropoff location IDs. We'll use only these two features for our model. 

* Turn the dataframe into a list of dictionaries (remember to re-cast the ids to strings - otherwise it will 
  label encode them)
* Fit a dictionary vectorizer 
* Get a feature matrix from it

What's the dimensionality of this matrix (number of columns)?

* 2
* 155
* 345
* 515
* 715

```python
x_dict = df_norm[['PULocationID', 'DOLocationID']].astype(str).to_dict(orient='records')
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
dv = DictVectorizer()
m = dv.fit_transform(x_dict)
```

```python
m
```

## Q5. Training a model

Now let's use the feature matrix from the previous step to train a model. 

* Train a plain linear regression model with default parameters, where duration is the response variable
* Calculate the RMSE of the model on the training data

What's the RMSE on train?

* 3.64
* 7.64
* 11.64
* 16.64

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
model = LinearRegression()
```

```python
y_train = df_norm['duration']
```

```python
model.fit(m, y_train)
```

```python
y_pred = model.predict(m)
```

```python
rmse1 = np.sqrt(mean_squared_error(y_train, y_pred))
rmse1
```

## Q6. Evaluating the model

Now let's apply this model to the validation dataset (February 2023). 

What's the RMSE on validation?

* 3.81
* 7.81
* 11.81
* 16.81


```python
dfval = pd.read_parquet("data/yellow_tripdata_2023-02.parquet")
```

```python
dfval.shape[0]
```

```python
dfval['duration'] = (dfval['tpep_dropoff_datetime'] - dfval['tpep_pickup_datetime']).dt.total_seconds() / 60
```

```python
dfval_norm = dfval[dfval['duration'].between(1, 60)]
```

```python
xval_dict = dfval_norm[['PULocationID', 'DOLocationID']].astype(str).to_dict(orient='records')
```

```python
m_val = dv.transform(xval_dict)
```

```python
yval = dfval_norm['duration']
```

```python
yval_pred = model.predict(m_val)
```

```python
rmse2 = np.sqrt(mean_squared_error(yval, yval_pred))
rmse2
```
